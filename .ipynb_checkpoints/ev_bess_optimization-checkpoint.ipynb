```python
import pandas as pd
import numpy as np
import cvxpy as cp
import gurobipy as gp
```

```python
# -----------------------------
# CONFIG PARAMETERS
# -----------------------------
# net metering tariff structure to optimize for
nem = "NBT" # NEM2 or NBT

# Rate Plan
rate_plan = "EV2-A"

# Battery
bess_capacity = 12
bess_max_power = 5
bess_eff = 0.95
bess_min_soc = 0.1 * bess_capacity
bess_max_soc = 0.9 * bess_capacity

# EV
ev_capacity = 40
ev_max_power = 5
ev_eff = 0.95
ev_min_soc = 0.1 * ev_capacity
ev_max_soc = 0.9 * ev_capacity
ev_daily_miles = 30
ev_mpg = 3
ev_daily_kwh = ev_daily_miles / ev_mpg  # kWh/day
```

```python
# TOU definition for PG&E EV-2A rate plan
def define_tou(h):
    if 16 <= h <= 20:
        return "peak"
    elif (21 <= h <= 23 or h == 15):
        return "part-peak"
    else:
        return "off-peak"
```

```python
# -----------------------------
# DATA READ / WRANGLING
# -----------------------------
path = <path>
data = pd.read_csv(path + "csv_in/input_data_residential.csv")
rates = pd.read_csv(path + "csv_in/pge_rate_plans.csv")

T = len(data)  # should be 8760
data["datetime"] = pd.to_datetime(data[["year", "month", "day", "hour"]])
data["weekday"] = pd.Series(np.where(data["datetime"].dt.dayofweek >= 5, "weekend", "weekday"))

# merge rate plans into data
data["season"] = np.where(data["month"].between(6, 9), "summer", "winter")
data["tou"] = data["hour"].apply(define_tou)
rates_ev = rates[rates["rate_plan"] == rate_plan].copy().drop(columns = ["rate_plan"])
data = data.merge(
    rates_ev,
    on=["season", "tou"],
    how="left"
)

# define EV availability (1=available)
weekday = data["weekday"].eq("weekday")
weekend = data["weekday"].eq("weekend")
data["ev_available"] = (
    (weekday & ((data["hour"] >= 17) | (data["hour"] <= 8))) |
    (weekend & ((data["hour"] >= 21) | (data["hour"] <= 16)))
).astype(int)

# export CSV data for checking
# data.to_csv(path + "csv_out/py_optim_data.csv", index=False)
```

```python
# extract vectors required for optimization
load = data["customer_load"].values
solar = data["pv_output"].values
import_price = data["pge_rate"].values
if nem == "NBT":
    export_price = data["fit_nbt"].values
else:
    export_price = data["pge_rate"].values
ev_available = data["ev_available"].values
```

```python
# -----------------------------
# DECISION VARIABLES
# -----------------------------
grid_import = cp.Variable(T, nonneg=True)
grid_export = cp.Variable(T, nonneg=True)

bess_charge = cp.Variable(T, nonneg=True)
bess_discharge = cp.Variable(T, nonneg=True)
bess_soc = cp.Variable(T)

ev_charge = cp.Variable(T, nonneg=True)
ev_discharge = cp.Variable(T, nonneg=True)
ev_soc = cp.Variable(T)
```

```python
# -----------------------------
# OBJECTIVE FUNCTION
# -----------------------------
cost = cp.sum(
    cp.multiply(grid_import, import_price)
    - cp.multiply(grid_export, export_price))

# 5c/kWh penalty to avoid EV battery degradation
cost += 0.05 * cp.sum(ev_discharge) 
```

```python
# -----------------------------
# CONSTRAINTS
# -----------------------------
constraints = []

# Energy balance: customer load can be supplied from grid, solar, bess, or EV
for t in range(T):
    constraints += [
        load[t] == 
        grid_import[t] - grid_export[t]
        + solar[t]
        + bess_discharge[t] - bess_charge[t]
        + ev_discharge[t] - ev_charge[t]
    ]

# Can't export more energy than is physically available
for t in range(T):
    constraints += [
        grid_export[t] <= solar[t] + bess_discharge[t] + ev_discharge[t]
    ]

# Grid exports cannot exceed 100% of load in each month (required for NEM2)
for m in range(12):
    idx = (data["month"] == m+1)
    constraints += [
        cp.sum(grid_export[idx]) <= 1.0 * np.sum(load[idx])
    ]

# BESS constraints
for t in range(T):
    constraints += [
        bess_charge[t] <= bess_max_power, # constraint on rate of dis/charge
        bess_discharge[t] <= bess_max_power,
        bess_soc[t] >= bess_min_soc, # constraint on state of charge (SoC)
        bess_soc[t] <= bess_max_soc
    ]

    if t == 0:
        constraints += [
            bess_soc[t] == bess_capacity * 0.5 # initial BESS SoC = 50% of cap
        ]
    else:
        constraints += [
            bess_soc[t] == bess_soc[t-1]
            + bess_eff * bess_charge[t]
            - bess_discharge[t] / bess_eff
        ]

# EV constraints
# Daily EV driving consumption
ev_drive = np.zeros(T)
for t in range(T):
    if ev_available[t] == 0:
        ev_drive[t] = ev_daily_kwh / 8 # driving pattern unknown hence uniformly spread
    else:
        ev_drive[t] = 0

for t in range(T):
    constraints += [
        ev_charge[t] <= ev_max_power * ev_available[t],
        ev_discharge[t] <= ev_max_power * ev_available[t],
        ev_soc[t] >= ev_min_soc,
        ev_soc[t] <= ev_max_soc
    ]

    if t == 0:
        constraints += [
            ev_soc[t] == ev_capacity * 0.5 # initial EV SoC = 50% of cap
        ]
    else:
        constraints += [
            ev_soc[t] == ev_soc[t-1]
            + ev_eff * ev_charge[t]
            - ev_discharge[t] / ev_eff
            - ev_drive[t]
        ]
```

```python
# -----------------------------
# SOLVE
# -----------------------------
prob = cp.Problem(cp.Minimize(cost), constraints)
prob.solve(solver=cp.GUROBI, verbose=False)  # or SCS solver

print("Optimized for", nem)
print("Total annual cost: $", round(prob.value, 2))
```

```python
# export to CSV
results = pd.DataFrame({
    "datetime": data["datetime"].values,
    "pge_rate": data["pge_rate"].values,
    "fit_nbt": data["fit_nbt"].values,
    "ev_available": data["ev_available"].values,
    "grid_import": grid_import.value,
    "grid_export": grid_export.value,
    "bess_soc": bess_soc.value,
    "ev_soc": ev_soc.value,
})

# total cost under NEM2 & NBT ignoring EV over-charging penalty
results["cost_nem2"] = (results["grid_import"] - results["grid_export"]) * results["pge_rate"]
results["cost_nbt"] = results["grid_import"] * results["pge_rate"] - results["grid_export"] * results["fit_nbt"]

results.to_csv(path + "csv_out/py_optim_output_" + nem + ".csv", index=False)
```
